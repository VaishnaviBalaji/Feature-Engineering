```python
import pandas as pd
df=pd.read_csv('spinal.csv')
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(labels=['labels'], axis=1),
    df['labels'],
    test_size=0.3,
    random_state=0)
```

```python
X_train
```

```python

from sklearn.feature_selection import mutual_info_regression
# determine the mutual information
mutual_info = mutual_info_regression(X_train.fillna(0), y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
#mutual_info.index = X_train.columns
#mutual_info.sort_values(ascending=False)
```

```python
mutual_info.plot.bar(figsize=(85,45))
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
## Selecting the top 20 percentile
selected_top_columns = SelectPercentile(mutual_info_regression, percentile=10)
selected_top_columns.fit(X_train.fillna(0), y_train)
```
