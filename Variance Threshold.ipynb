```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.feature_selection import VarianceThreshold
```

```python
df=pd.read_csv('spinal.csv') #jointed the data,feature,labels dataset from moodle into a single spinal.csv
```

```python
df.shape
```

```python
df.head()
```

```python
#Here labels is the dependent feature, so kick it out from the data and give only the independent data to x
x=df.drop(labels=['labels'],axis=1)
#dependent feature in y
y=df['labels']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    df.drop(labels=['labels'],axis=1),
    df['labels'],
    test_size=0.3,
    random_state=0)

x_train.shape,x_test.shape
#70-30 rule, training with 70% of the data and testing with 30% of the data
```

```python
#Here given variance Threshold value is 0.1 that is variance for those columns is 0.1 
var_thres=VarianceThreshold(threshold=0.1)
var_thres.fit(x_train)#Performing the feature selection model on our training dataset
```

```python
sum(var_thres.get_support())
#gives the sum of good columns (which has variance > 0.1)
```

```python
var_thres.get_support()
#Here True is good columns :) and false is bad columns :(
```

```python
low_variance_columns=[column for column in x_train.columns
                    if column not in x_train.columns[var_thres.get_support()]]
#Here we all the bad(false) columns in var_thres is fed to low_variance_columns
#so low_variance_columns will now have all the bad columns 
```

```python
print(len(low_variance_columns))
#len of all bad columns witht the variance 0.1
```

```python
for column in low_variance_columns:
    print(column)

#Label names of the bad columns
```

```python
#Drop the bad columns...BOOM!!!
x_train.drop(low_variance_columns,axis=1)
```

```python
#Summary:
#By using the VarianceThreshold with the threshold value 0.1, our data set with 432 columns dropped to 139...!!!
#We kindof removed 293 features which has low variance
#That pretty good, but this is medical data(which is pretty sensitve, since it deals with human life) so I'm not sure if we can increase the threshold value even more and reduce even more features....Let's see!!! 
```
